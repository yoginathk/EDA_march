**Standardization**

- It is the technique scale all the data under one scale
- Different column has different values also different units
- Some columns values has bigger values, some column values has lesser values
- So it is important to keep all the values under one scale.

- we have two methods are there...
    - standardization 
    - Normalization
    - 
- Standarization
   - Z -score
   - Z score varies values from -3 to 3
   
   <strong>z = (x – μ) / σ</strong>

- Normalization
     - min max scalar

**task**
- step-1: take prevailing wage column
- step-2 : calculate the mean value prevailing wage
- step-3: calculate the standard deviation of prevailing wage
- step-4: calculate the Nr: step-1 - step-2
- Step-5: divide the step4/step3

```python
import pandas as pd
visa_df = pd.read_csv("../datafiles/Visadataset.csv")
```

```python
 
mean=visa_df['prevailing_wage'].mean()
std=visa_df['prevailing_wage'].std()
Nr = visa_df['prevailing_wage'] - mean
visa_df['prevailing_wage_z' ] = Nr/std
visa_df
```

**Task-2**

**idxmax- idxmin**


```python
maxx =  visa_df['prevailing_wage'].idxmax(), visa_df['prevailing_wage_z'].idxmax()
minn = visa_df['prevailing_wage'].idxmin(), visa_df['prevailing_wage_z'].idxmin()
```

```python
from sklearn.preprocessing import StandardScaler
```
